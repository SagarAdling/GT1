 HISTOGRAM:-
    It is used to graphing the frequency over a distribution it is a very uselful graph in the analytics world and can many useful
    insight from the data. visually all the bars touchting each other with no space between them,
    it is a type of bar plot where the x axis represents the bin ranges while the y axis gives infomation about frequency.....

```python
x= df['total_bill']
plt.hist(x,bins=25, color='orange',edgecolor='blue',linestyle='--', alpha = 0.5)
plt.title("Tips Dataset")
plt.ylabel('frequency')
plt.xlabel('Total Bill')
plt.show()
```

# Different types of graph
count plot :
A count plot is helpful when dealing with categorical values. it is used to plot the frequency of the different categories.

```python
df['gender'].value_counts()
```

```python
sns.countplot(x = 'gender', data = df)
```

```python
sns.countplot(x = 'gender', hue = 'day', data = df, palette = "set2")
```

KDE plot
A kernel density estimate plot is used to plot the distribution of countinous data.

```python
df2 = sns.load_dataset('iris')
df2.head(5)
```

```python
# setting the color of the plot 
colors = {'setosa': 'b','versicolor': 'g', 'vriginica' : 'r'}
# creating the subplots
fig, ax = plt.subplots()
#plot each data.point
for i in range(len(df2['sepal_length'])):
    #ploting different classes with different color
    ax.scatter(df2['petal_length                    '][i],df2['petal_width'][i],
               color=colors[df2['species'][i]])
```

```python
sns.scatterplot(x = 'sepal_length',y = 'petal_length',data = df2,hue = 'species')
plt.show()
```

```python
# pair  plots
sns.pairplot(df2,hue = 'species')
```
